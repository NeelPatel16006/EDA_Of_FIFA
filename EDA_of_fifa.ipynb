```python
# Top 10 countrys fifa players highest

fifa['nationality'].value_counts()[:10]
```

```python
# plot with bar plot

plt.figure(figsize=(7,5))
plt.bar(list(fifa['nationality'].value_counts()[:10].keys()),list(fifa['nationality'].value_counts()[:10]),color='orange')
plt.xticks(rotation='vertical')
plt.show()
```

```python
# player salary

player_salary=fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

```python
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
player_salary.index=np.arange(1,len(player_salary['short_name'])+1)
player_salary.head()
```

```python
# Top 10 players whos salary high

player_salary.iloc[:10,:]
```

```python
# plot using bar plot

plt.bar(list(player_salary['short_name'])[:5],list(player_salary['wage_eur'])[:5],color=['red','green','orange','yellow','black'])
plt.xticks(rotation='vertical')
plt.show()
```
